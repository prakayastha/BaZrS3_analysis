```python
from scipy import constants
import matplotlib.pyplot as plt
import numpy as np
```

```python
T = []
G = []
with open('gibbs-temperature.dat') as f:
    for line in f:
        #print(line.split())
        T.append(float(line.split()[0]))
        G.append(float(line.split()[1])*constants.physical_constants["electron volt-joule relationship"][0]*constants.N_A*0.001)
#print(T)
#print(G)
```

```python
plt.plot(T,G,"--", label='Ba$_2$ZrS$_4$')
plt.ylabel(r"$\Delta G$ (kJ mol$^{-1}$)")
plt.xlim([0,2000])
plt.ylim([min(G),max(G)])
plt.yticks(np.arange(min(G), max(G)+1, 100))
plt.xlabel("Temperature (K)")
plt.grid(alpha=0.4)
plt.tight_layout()
plt.legend()
plt.savefig('G_T_ternary_phases')
```

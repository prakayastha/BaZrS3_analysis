```python
import numpy as np 
temperatures = []
BaZrS3_GFE = []
Ba4Zr3S10_GFE = []
Ba2ZrS4_GFE = []
ZrS2_GFE = []
Ba3Zr2S7_GFE = []

with open('BZS_4310_LD_214_ZrS2_327') as f:
    for line in f:
        #print(line)
        line = line.split()
        temperature = float(line[0])
        temperatures.append(temperature)
        BaZrS3 = float(line[1])
        BaZrS3_GFE.append(BaZrS3)
        Ba4Zr3S10 = float(line[3])
        Ba4Zr3S10_GFE.append(Ba4Zr3S10)      
        Ba2ZrS4 = float(line[5])
        Ba2ZrS4_GFE.append(Ba2ZrS4)  
        ZrS2 = float(line[7])
        ZrS2_GFE.append(ZrS2)
        Ba3Zr2S7 = float(line[9])
        Ba3Zr2S7_GFE.append(Ba3Zr2S7)
#print(line)
BaZrS3_GFE = np.array(BaZrS3_GFE)
Ba2ZrS4_GFE = np.array(Ba2ZrS4_GFE)
Ba3Zr2S7_GFE = np.array(Ba3Zr2S7_GFE)
Ba4Zr3S10_GFE = np.array(Ba4Zr3S10_GFE)
```

```python
from scipy import constants 

eV2kJmol = (constants.physical_constants["electron volt-joule relationship"][0] * constants.N_A/1000)
```

## 2BaZrS3 -> Ba2ZrS4 + ZrS2

```python
reactants = BaZrS3_GFE
products = Ba2ZrS4_GFE+ZrS2_GFE
#print(reactants)
#print(products)
res1 = (products-reactants/2)/2*eV2kJmol
print(res1)
```

## 4BaZrS3 -> Ba4Zr3S10 + ZrS2

```python
reactants = BaZrS3_GFE
products = Ba4Zr3S10_GFE+ZrS2_GFE
#print(reactants)
#print(products)
res2 = (products-reactants)*eV2kJmol
print(res2)
```

## 3BaZrS3 -> Ba3Zr2S7 + ZrS2

```python
reactants = BaZrS3_GFE
products = Ba3Zr2S7_GFE+ZrS2_GFE
#print(reactants)
#print(products*4/3)
res3 = ((products*4/3)-reactants)*eV2kJmol
print(res3)
```

```python
import matplotlib.pyplot as plt
plt.plot(temperatures,res1,"--", label="214")
plt.plot(temperatures,res2,"--", label="4310")
plt.plot(temperatures,res3,"-", label="327")
plt.ylabel(r"$\Delta G$ (kJ mol$^{-1}$)")
#plt.yticks([2,4,6,8,10,12,14,16])
plt.xlabel("Temperature")
```

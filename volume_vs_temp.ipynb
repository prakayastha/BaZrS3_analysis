```python
import os 
import matplotlib.pyplot as plt
```

```python
temperature = []
for temp in range(1,21):
    temperature.append(temp*100)
print(temperature)
```

```python
with open('QHA/Ba2ZrS4_I4_mmm/helmholtz-volume_fitted.dat') as f:
    while '# Minimas' not in f.readline():
        continue
    lines = f.readlines()
    volume_214 = []
    for i in lines:
        if len(i.split()) > 1:
            volume_214.append(float((i.split()[0])))
    print(volume_214)
```

```python
with open('QHA/Ba4Zr3S10_I4_mmm/helmholtz-volume_fitted.dat') as f:
    while '# Minimas' not in f.readline():
        continue
    lines = f.readlines()
    volume_4310 = []
    for i in lines:
        if len(i.split()) > 1:
            volume_4310.append(float((i.split()[0])))
    print(volume_4310)
```

```python
plt.plot(temperature,volume_214)
plt.plot(temperature,volume_4310)
```
